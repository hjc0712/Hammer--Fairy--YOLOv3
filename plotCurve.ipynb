```python
mAP = []
precision = []
recall = []
f1 = []
train_mAP = []

```

```python
with open('auto_Test_final3.txt') as fp:
    lines = fp.readlines()
    for line in lines:
        if(line.startswith("mAP")):
            mAP.append(float(line[4:]))
        elif(line.startswith("precision")):
            precision.append(float(line[10:]))
        elif(line.startswith("recall")):
            recall.append(float(line[7:]))
        elif(line.startswith("f1")):
            f1.append(float(line[3:]))
```

```python
import matplotlib.pyplot as plt
```

```python
x = [i*5 for i in range(1,59)]
y = mAP

fig1, ax1 = plt.subplots()
plt.plot(x,y)
plt.xlabel("iteration")
plt.ylabel("mAP")
plt.title("mAP on test set")
plt.savefig("mAP.png")

y = precision
fig2, ax2 = plt.subplots()
plt.plot(x,y)
plt.xlabel("iteration")
plt.ylabel("Precision at confidence threshold 0.5")
plt.title("Precision on test set with confidence threshold at 0.5 ")
plt.savefig("Precision.png")

y = recall
fig2, ax2 = plt.subplots()
plt.plot(x,y)
plt.xlabel("iteration")
plt.ylabel("Recall at confidence threshold 0.5")
plt.title("Recall on test set with confidence threshold at 0.5 ")
plt.savefig("Recall.png")

y = f1
fig2, ax2 = plt.subplots()
plt.plot(x,y)
plt.xlabel("iteration")
plt.ylabel("F1 at confidence threshold 0.5")
plt.title("F1 on test set with confidence threshold at 0.5 ")
plt.savefig("F1.png")
```
